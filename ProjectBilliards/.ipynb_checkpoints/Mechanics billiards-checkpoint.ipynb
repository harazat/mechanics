```python
import math
import random
import tkinter as tk
from turtle import Turtle, Screen


class CircularBilliardGUI(tk.Tk):
    def __init__(self):
        super().__init__()

        self.title("Circular Billiard Simulation")
        self.geometry("1000x1000")

        self.reflections = 30
        self.path = None
        self.reversed_path = None

        self.create_widgets()

        #some codes for GUI taken from  https://www.youtube.com/watch?v=3UVjr8Onll8
        # and at https://docs.python.org/3/library/turtle.html
    def create_widgets(self):
        self.canvas = tk.Canvas(self, width=1000, height=1000)
        self.canvas.pack(side=tk.TOP, pady=10)

     
        self.start_button = tk.Button(self, text="Start Simulation", command=self.start_simulation)
        self.start_button.place(relx=0.5, rely=0.5, anchor=tk.CENTER)

        self.reflections_entry = tk.Entry(self, width=5)
        self.reflections_entry.insert(tk.END, str(self.reflections))
        self.reflections_entry.place(relx=0.5, rely=0.55, anchor=tk.CENTER)

        self.reverse_button = tk.Button(self, text="Reverse Path", command=self.reverse_path)
        self.reverse_button.place(relx=0.5, rely=0.6, anchor=tk.CENTER)


        self.reversibility_button = tk.Button(self, text="Test Reversibility", command=self.test_reversibility)
        self.reversibility_button.place(relx=0.5, rely=0.65, anchor=tk.CENTER)

    def start_simulation(self):
        try:
            self.reflections = int(self.reflections_entry.get())
            if self.reflections < 0:
                raise ValueError
        except ValueError:
            tk.messagebox.showerror("Invalid Input", "Number of reflections must be a positive integer.")
            return

        self.canvas.delete("all")

        screen = Screen()
        screen.clear()


        turtle = Turtle()
    
        turtle.speed(10)  


        turtle.penup()
        turtle.goto(0, -200)
        turtle.pendown()
        turtle.circle(200)
        turtle.penup()

       
        angle = random.uniform(0, 2 * math.pi)
        radius = random.uniform(0, 200)
        x = radius * math.cos(angle) / 200 
        y = radius * math.sin(angle) / 200  

      
        momentum_angle = random.uniform(0, 2 * math.pi)
        px = math.cos(momentum_angle)
        py = math.sin(momentum_angle)

      
        turtle.goto(x * 200, y * 200)  
        turtle.pendown()
        turtle.color("red")
        turtle.dot(5)

     
        self.path = []
        self.path.append((x, y))

        for reflection in range(self.reflections):
            a = px ** 2 + py ** 2
            b = 2 * (x * px + y * py)
            c = x ** 2 + y ** 2 - 1 
            discriminant = b ** 2 - 4 * a * c

            if discriminant < 0:
                discriminant = 0 

            t = (-b + math.sqrt(discriminant)) / (2 * a)
            x += t * px
            y += t * py

            px, py = (y ** 2 - x ** 2) * px - 2 * x * y * py, -2 * x * y * px + (x ** 2 - y ** 2) * py

            self.path.append((x, y))
            turtle.goto(x * 200, y * 200) 
            turtle.pendown()
            turtle.color("red")
            turtle.dot(5)

           
            print(f"Reflection {reflection + 1}: ({x:.2f}, {y:.2f})")

     
        screen.update()

    def reverse_path(self):
        if self.path is None:
            tk.messagebox.showerror("No Path", "Please start the simulation first.")
            return

        
        self.reversed_path = self.path[::-1]

      
        turtle = Turtle()
        turtle.width(2)
        turtle.speed(10)  

        
        for i in range(len(self.reversed_path) - 1):
            start = self.reversed_path[i]
            end = self.reversed_path[i + 1]
            turtle.goto(start[0] * 200, start[1] * 200)  
            turtle.pendown()
            turtle.color("purple")
            turtle.dot(5)
            turtle.goto(end[0] * 200, end[1] * 200)  
            turtle.penup()

     
        turtle.screen.update()

    def test_reversibility(self):
        if self.path is None:
            tk.messagebox.showerror("No Path", "Please start the simulation first.")
            return

        if self.reversed_path is None:
            tk.messagebox.showerror("No Reversed Path", "Please reverse the path first.")
            return

       
        deviation = 0
        for i in range(len(self.path)):
            x1, y1 = self.path[i]
            x2, y2 = self.reversed_path[i]
            deviation += math.sqrt((x1 - x2) ** 2 + (y1 - y2) ** 2)

      
        threshold = 0.01

        if deviation <= threshold:
            tk.messagebox.showinfo("Reversibility Test", f"The paths coincide within the specified deviation ({threshold}).")
        else:
            tk.messagebox.showinfo("Reversibility Test", f"The paths deviate more than the specified deviation ({threshold}).")


if __name__ == "__main__":
    app = CircularBilliardGUI()
    app.mainloop()
```

```python
import math
import random
import tkinter as tk
from turtle import Turtle, Screen


class CircularBilliardGUI(tk.Tk):
    def __init__(self):
        super().__init__()

        self.title("Circular Billiard Simulation")
        self.geometry("800x600")

        self.reflections = 10
        self.path = None
        self.reversed_path = None

        self.create_widgets()

    def create_widgets(self):
    
        self.canvas = tk.Canvas(self, width=600, height=600)
        self.canvas.pack(side=tk.TOP, pady=10)

       
        self.start_button = tk.Button(self, text="Start Simulation", command=self.start_simulation)
        self.start_button.place(relx=0.5, rely=0.5, anchor=tk.CENTER)

      
        self.reflections_entry = tk.Entry(self, width=5)
        self.reflections_entry.insert(tk.END, str(self.reflections))
        self.reflections_entry.place(relx=0.5, rely=0.55, anchor=tk.CENTER)

        self.reverse_button = tk.Button(self, text="Reverse Path", command=self.reverse_path)
        self.reverse_button.place(relx=0.5, rely=0.6, anchor=tk.CENTER)

        self.screen = Screen()
        self.turtle = Turtle()
        self.turtle.speed(5)  

      
        self.turtle.penup()
        self.turtle.goto(0, -200)
        self.turtle.pendown()
        self.turtle.circle(200)
        self.turtle.penup()

    def start_simulation(self):
        try:
            self.reflections = int(self.reflections_entry.get())
            if self.reflections < 0:
                raise ValueError
        except ValueError:
            tk.messagebox.showerror("Invalid Input", "Number of reflections must be a positive integer.")
            return

        
        self.canvas.delete("all")

        
        self.path = []

        
        self.turtle.clear()

        
        angle = random.uniform(0, 2 * math.pi)
        radius = random.uniform(0, 200)
        x = radius * math.cos(angle) / 200  
        y = radius * math.sin(angle) / 200  
        
        momentum_angle = random.uniform(0, 2 * math.pi)
        px = math.cos(momentum_angle)
        py = math.sin(momentum_angle)

        
        self.turtle.goto(x * 200, y * 200)  
        self.turtle.pendown()
        self.turtle.color("blue")
        self.turtle.dot(5)

      
        self.path.append((x, y))

        gravitational_force = 10

        for reflection in range(self.reflections):
            a = px ** 2 + py ** 2
            b = 2 * (x * px + y * py)
            c = x ** 2 + y ** 2 - 1  
            discriminant = b ** 2 - 4 * a * c

            if discriminant < 0:
                discriminant = 0

            t = (-b + math.sqrt(discriminant)) / (2 * a)
            x += t * px
            y += t * py

            angle_incidence = math.atan2(py, px)

          
            angle_reflection = math.atan2(y, x) - 2 * angle_incidence

         
            px = (y ** 2 - x ** 2) * math.cos(angle_reflection) - 2 * x * y * math.sin(angle_reflection)
            py = -2 * x * y * math.cos(angle_reflection) + (x ** 2 - y ** 2) * math.sin(angle_reflection)

           
            py -= gravitational_force / 200

            self.path.append((x, y))
            self.turtle.goto(x * 200, y * 200)  
            self.turtle.pendown()
            self.turtle.color("blue")
            self.turtle.dot(5)

           
            print(f"Reflection {reflection + 1}: ({x:.2f}, {y:.2f})")

       
        self.screen.update()

    def reverse_path(self):
        if self.path is None:
            tk.messagebox.showerror("No Path", "Please start the simulation first.")
            return

     
        self.reversed_path = self.path[::-1]

        
        turtle_reversed = Turtle()
        turtle_reversed.speed(2.5)  

      
        for i in range(len(self.reversed_path) - 1):
            start = self.reversed_path[i]
            end = self.reversed_path[i + 1]
            turtle_reversed.goto(start[0] * 200, start[1] * 200)  
            turtle_reversed.pendown()
            turtle_reversed.color("red")
            turtle_reversed.dot(5)
            turtle_reversed.goto(end[0] * 200, end[1] * 200)  
            turtle_reversed.penup()

    
        turtle_reversed.screen.update()


if __name__ == "__main__":
    app = CircularBilliardGUI()
    app.mainloop()
```

```python
import math
import random
import tkinter as tk
from tkinter import messagebox
from turtle import Turtle, Screen


class StadiumBilliardGUI(tk.Tk):
    def __init__(self):
        super().__init__()

        self.title("Stadium Billiard Simulation")
        self.geometry("800x600")

        self.reflections = 10
        self.path = None

        self.create_widgets()

    def create_widgets(self):
   
        self.canvas = tk.Canvas(self, width=600, height=400)
        self.canvas.pack(side=tk.TOP, pady=10)

     
        self.start_button = tk.Button(self, text="Start Simulation", command=self.start_simulation)
        self.start_button.place(relx=0.5, rely=0.5, anchor=tk.CENTER)

        
        self.reflections_entry = tk.Entry(self, width=5)
        self.reflections_entry.insert(tk.END, str(self.reflections))
        self.reflections_entry.place(relx=0.5, rely=0.55, anchor=tk.CENTER)

    def start_simulation(self):
        try:
            self.reflections = int(self.reflections_entry.get())
            if self.reflections < 0:
                raise ValueError
        except ValueError:
            tk.messagebox.showerror("Invalid Input", "Number of reflections must be a positive integer.")
            return

      
        self.canvas.delete("all")

        
        screen = Screen()
        screen.tracer(0)  

        turtle = Turtle()
        turtle.speed(0)  


        window_size = 600
        turtle.penup()
        turtle.goto(-window_size / 2, -window_size / 2)
        turtle.pendown()

        
        scale_factor = window_size / 2 
        turtle.circle(scale_factor, 90)
        turtle.forward(window_size)
        turtle.circle(scale_factor, 90)
        turtle.forward(window_size)
        turtle.penup()

      
        x = random.uniform(-1 + L, 1 - L)
        y = random.uniform(-1, 1)

     
        px = random.uniform(-1, 1)
        py = random.uniform(-1, 1)
        norm = math.sqrt(px**2 + py**2)
        px /= norm
        py /= norm

     
        turtle.goto(x * scale_factor, y * scale_factor)
        turtle.pendown()
        turtle.color("blue")
        turtle.dot(5)

     
        self.path = []
        self.path.append((x, y))

        for reflection in range(self.reflections):
            if y > 0:
                if abs(x) < L / 2:
                    intersect_y = 1
                    intersect_x = x + (intersect_y - y) * (px / py)
                else:
                    intersect_x = L / 2
                    intersect_y = math.sqrt(1 - (intersect_x - L / 2)**2)
            else:
                if abs(x) < L / 2:
                    intersect_y = -1
                    intersect_x = x + (intersect_y - y) * (px / py)
                else:
                    intersect_x = -L / 2
                    intersect_y = -math.sqrt(1 - (intersect_x + L / 2)**2)

            self.path.append((intersect_x, intersect_y))


            if intersect_y == 1 or intersect_y == -1:
                px, py = px, -py  
            else:
                px = (intersect_y**2 - (intersect_x - L / 2)**2) * px - 2 * (intersect_x - L / 2) * intersect_y * py
                py = -2 * (intersect_x - L / 2) * intersect_y * px + ((intersect_x - L / 2)**2 - intersect_y**2) * py

            x = intersect_x
            y = intersect_y


            turtle.goto(x * scale_factor, y * scale_factor)
            turtle.dot(5)

        turtle.hideturtle()

        self.perform_pseudo_random_test()

        screen.exitonclick()

    def perform_pseudo_random_test(self):
        x_values = [point[0] for point in self.path]
        bin_count = 10 

        bin_counts = [0] * bin_count

        for value in x_values:
            bin_index = int((value + 1) * bin_count / 2)
            bin_counts[bin_index] += 1

        expected_entries = self.reflections / bin_count
        mean_entries = sum(bin_counts) / bin_count
        variance = sum((count - expected_entries) ** 2 for count in bin_counts) / bin_count

        tk.messagebox.showinfo("Pseudo-random Test Results",
                               f"Mean number of entries in the bins: {mean_entries:.2f}\n"
                               f"Variance of the number of entries: {variance:.2f}")


if __name__ == "__main__":
    L = 0.5  

    app = StadiumBilliardGUI()
    app.mainloop()
```

```python
import math
import random
import numpy as np
import matplotlib.pyplot as plt
import tkinter as tk

def calculate_intersection(x, y, px, py, L):
   
    if abs(x) >= L/2:
        py = -py  
    x += px
    y += py

    return x, y, px, py


def simulate_stadium_billiard(n, L):

    r = 2
    epsilon = 1e-6

   
    x = random.uniform(-r, r)
    y = random.uniform(-r, r)
    px = random.uniform(-1, 1)
    py = math.sqrt(1 - px**2) if abs(px) < 1 - epsilon else random.choice([-1, 1])

    reflection_points = []

    for _ in range(n):
        if math.sqrt(x**2 + y**2) > r:
            break

        x, y, px, py = calculate_intersection(x, y, px, py, L)
        reflection_points.append((x, y, px, py))

    return reflection_points

def test_uniformity(reflection_points, M):
 
    bin_counts, _ = np.histogram(reflection_points, bins=M)

  
    mean_count = np.mean(bin_counts)
    var_count = np.var(bin_counts)

    return bin_counts, mean_count, var_count

def draw_path(reflection_points):
   
    root = tk.Tk()
    root.title("Stadium Billiard Path")

  
    canvas_width = 400
    canvas_height = 400
    canvas = tk.Canvas(root, width=canvas_width, height=canvas_height)
    canvas.pack()

  
    for i in range(len(reflection_points)):
        x, y, px, py = reflection_points[i]
        color = "blue" if px > 0 else "red"
        canvas.create_oval(x * canvas_width - 2, y * canvas_height - 2, x * canvas_width + 2, y * canvas_height + 2, fill=color)

 
    root.mainloop()

n = 200  
L = 0.5
M = 10


reflection_points = simulate_stadium_billiard(n, L)




bin_counts, mean_count, var_count = test_uniformity(reflection_points, M)

fig = plt.figure()
ax = fig.add_subplot(111)
ax.bar(range(M), bin_counts)
ax.set_xlabel("Bin")
ax.set_ylabel("Count")
ax.set_title("Histogram of Bin Counts")

plt.show()

print("Mean Count:", mean_count)
print("Variance of Count:", var_count)

draw_path(reflection_points)
```
